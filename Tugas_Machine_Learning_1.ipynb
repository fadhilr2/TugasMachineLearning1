```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fadhilr2/TugasMachineLearning1/refs/heads/main/dataset/asylum_seekers.csv')
```

# Data cleaning

```python
df = df.fillna(value = "")
```

```python
df.rename(columns={'Origin':'country_of_origin', 'Applied during year':'applied_during_year'}, inplace=True)
```

```python
df['applied_during_year'] = pd.to_numeric(df['applied_during_year'], errors='coerce')
```

# Data Visualization

```python
df_2016 = df.query("Year == 2016")
```

```python
sums_2016 = df_2016.groupby(['country_of_origin'])[['applied_during_year']].agg('sum')
top_10_countries_of_origin_2016 = sums_2016.applied_during_year.sort_values(ascending=False)[:10]
chart_2016 = top_10_countries_of_origin_2016.plot.barh(
    figsize = [16, 8],
    fontsize = 14,
    title = '2016 Applications - Top 10 Countries of Origin',
    color = 'red')
chart_2016
```

Dari data ini dapat diketahui jumlah aplikasi permintaan untuk suaka(perlindungan yang diberikan oleh suatu negara kepada seseorang yang meninggalkan negara asalnya sebagai pengungsi politik) yang paling banyak berasal dari Suriah, berdasarkan data di tahun 2016.
